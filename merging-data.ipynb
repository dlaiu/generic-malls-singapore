# Importing the old data and standardising the formats

```python
import pandas as pd
```

```python
df = pd.read_csv('old_malls_matrix.csv')
```

```python
all_old_malls = {}

for col in df.columns[1:]:  # Iterate over each column except the first one
    temp_shops = []
    for index, row in df.iterrows():
        if row[col] == 1:
            # print(row[col])
            # print(row[df.columns[0]])
            temp_shops.append(row[df.columns[0]])
    
    all_old_malls[col] = {"shops": temp_shops}

all_old_malls
```

```python
import json

with open('all_old_malls.json', 'w') as f:
    json.dump(all_old_malls, f, indent=4)

print("Done!")
```

# Restructring Far East

```python
import json

with open('fareast_mall_shops.json', 'r') as f:
    fareast_mall_data = json.load(f)

restructured_data = {}
for key, value in fareast_mall_data.items():
    restructured_data[key] = {"shops": value}

restructured_data
```

```python
with open('fareast_mall_shops.json', 'w') as f:
    json.dump(restructured_data, f, indent=4)
```
