```python
import pandas as pd
from bs4 import BeautifulSoup
from playwright.async_api import async_playwright
import asyncio
```

# Scraping Capitaland Malls

```python
df_urls = pd.read_csv('malls_capitaland.csv')
df_urls
```

```python
playwright = await async_playwright().start()
browser = await playwright.chromium.launch(headless=False)
page = await browser.new_page()
```

```python
all_capitaland_mall_shops = {}
```

```python
for mall in df_urls['Malls'][1:4]:
    for url in df_urls[df_urls['Malls'] == mall]['Url']:
        current_url = url
        print(current_url)

    
    # current_url = df_urls[df_urls['Malls'] == mall]['Url'].to_string(index=False)

    await page.goto(current_url)

    if await page.locator("#truste-consent-button").is_visible():
        print("Element is visible.")
        await page.locator("#truste-consent-button").click()
    else:
        print("Element not found or not visible.")
    
    while True:
        try:
            await page.locator(".fn_see-more").click(timeout=5000) 
            await asyncio.sleep(4)
        except Exception as e:
            print("Failed with", e.message)
            break

    html = await page.content()
    soup = BeautifulSoup(html, 'html.parser')

    all_shops = []
    shops = soup.find_all(class_='title')
    for shop in shops:
        all_shops.append(shop.text)

    all_capitaland_mall_shops[mall] = all_shops

    await asyncio.sleep(2)

    print("done with", mall)

all_capitaland_mall_shops
```

```python
backup_mall_shops = all_capitaland_mall_shops
```

# Scraping Far East Malls

```python
df_fareast_urls = pd.read_csv('malls_fareast.csv')
df_fareast_urls
```

```python
playwright = await async_playwright().start()
browser = await playwright.chromium.launch(headless=False)
page = await browser.new_page()
```

```python
all_fareast_mall_shops = {}
```

```python
for mall in df_fareast_urls['Malls']:
    for url in df_fareast_urls[df_fareast_urls['Malls'] == mall]['Url']:
        current_url = url
        print(current_url)

    
    # current_url = df_urls[df_urls['Malls'] == mall]['Url'].to_string(index=False)

    await page.goto(current_url)

    if await page.locator("#truste-consent-button").is_visible():
        print("Element is visible.")
        await page.locator("#truste-consent-button").click()
    else:
        print("Element not found or not visible.")
    
    while True:
        try:
            await page.locator(".b-load-more-shops").click(timeout=5000) 
            await asyncio.sleep(4)
        except Exception as e:
            print("Failed with", e.message)
            break

    html = await page.content()
    soup = BeautifulSoup(html, 'html.parser')

    all_shops = []
    shops = soup.find_all(class_='b-store-title')
    for shop in shops:
        all_shops.append(shop.text)

    all_fareast_mall_shops[mall] = all_shops

    await asyncio.sleep(2)

    print("done with", mall)

all_fareast_mall_shops
```

```python
temp_fareast_malls = {}
```

```python
#cleaning the text
for mall in df_fareast_urls['Malls']:

    temp_shops = []

    for shop in all_fareast_mall_shops[mall]:
        temp_shops.append(shop.strip())

    temp_fareast_malls[mall] = temp_shops

temp_fareast_malls
```

```python
await browser.close()
```

```python
#export to json

with open('fareast_mall_shops.json', 'w') as f:
    json.dump(temp_fareast_malls, f, indent=4)

print(f'Dictionary has been exported as JSON file.')
```
