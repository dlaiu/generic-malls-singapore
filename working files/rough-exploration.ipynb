```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_shopping_malls_in_Singapore"
res = requests.get(url)

soup = BeautifulSoup(res.text, 'html.parser')

# print(soup)
```

```python
contents = soup.find_all(class_='div-col')
for content in contents:
    print(content.text)
```

```python
import pandas as pd

```

```python
data = ''
for content in contents:
    data = data + content.text

df = pd.DataFrame(data.split('\n'), columns=['Malls'])
df
```

```python
df = df.drop(index=0)
df
```

```python
df.to_csv('malls.csv', index=False)
```

```python
from playwright.async_api import async_playwright
```

```python
playwright = await async_playwright().start()
browser = await playwright.chromium.launch(headless=False)
page = await browser.new_page()
```

```python
await page.goto("https://www.capitaland.com/sg/malls/aperia/en/stores.html")
```

```python
import asyncio
```

```python
while True:
    try:
        await page.locator(".fn_see-more").click(timeout=5000) 
        await asyncio.sleep(1)
    except Exception as e:
        print("Failed with", e.message)
        break
```

```python
html = await page.content()
soup = BeautifulSoup(html, 'html.parser')
```

```python
all_shops = []
shops = soup.find_all(class_='title')
for shop in shops:
    all_shops.append(shop.text)

all_shops
```

```python
await browser.close()
```

```python
all_shops
```

```python
# Specify the text file name
file_name = "aperia_shops.txt"

# Open the file in write mode
with open(file_name, "w") as file:
    for shop in all_shops:
        # Write each element to the file followed by a newline character
        file.write(str(shop) + "\n")

print(f"Array saved to {file_name}")
```
